```python
def testgoogle(df):
    count=0
    for i in range(len(df)):
        hyperlink=tldextract.extract(df.loc[i]['url']).domain+'.'+ tldextract.extract(df.loc[i]['url']).suffix
        
        if 'google.com' in hyperlink:
            print(df.loc[i]['url'])
            df.at[i, 'cantina_new']=1
        else:
            df.at[i, 'cantina_new']=df.loc[i]['cantina']
    print(count)
```
